#### The model is basically a tree containing nodes and edges. There exist two types of nodes in the tree. Leaf nodes and decision nodes. Leaf nodes are the nodes which would be helpful in case of predicting (outputting the final value) while decision nodes will represent set of conditions that would help us to make a decision about the predicted value.

```python
class Node():
    def __init__(self, attribute=None, threshold=None, child_left=None, child_right=None, variance_red=None, leaf_value=None):
        
        # data members corresponding to decision nodes
        self.attribute = attribute
        self.threshold = threshold
        self.child_left = child_left
        self.child_right = child_right
        self.variance_red = variance_red
        
        #data member corresponding to a leaf node
        self.leaf_value = leaf_value
```

##### Kindly note: We have used the same defination of node for both the types of node. A decision node would have leaf_value = None while a leaf_node would have a numerical leaf_value. This difference would help us to differentiate between a leaf node and a decision node.

#### Class defination of a regression tree which will encapsulate all the functions and operation needed to construct a regression tree

```python
class RegressionTree():
    def __init__(self, minimum_samples=2, max_depth=2):
 
        self.root = None
        self.minimum_samples = minimum_samples
        self.max_depth = max_depth
        
    def construct_tree(self, dataset, current_depth=0):
       
        X, y = dataset[:,:-1], dataset[:,-1]
        number_datapoints, number_attributes = np.shape(X)
        current_best_feature = {}
        
        if number_datapoints >= self.minimum_samples and current_depth <= self.max_depth:
            current_best_feature = self.get_best_feature(dataset, number_datapoints, number_attributes)
            if current_best_feature["variance_reduced"]>0:
                subtree_left = self.construct_tree(current_best_feature["dataset_left"], current_depth+1)
                subtree_right = self.construct_tree(current_best_feature["dataset_right"], current_depth+1)
                return Node(current_best_feature["attribute"], current_best_feature["threshold"],subtree_left, subtree_right, current_best_feature["variance_reduced"])
        
        
        leaf_value = self.cal_leaf_node(y)
        return Node(leaf_value = leaf_value)
    
    def get_best_feature(self, dataset, number_datapoints, number_attributes):
        
        best_feature = {}
        best_feature["attribut"] = None
        best_feature["threshold"] = 0
        best_feature["dataset_left"] = None
        best_feature["dataset_right"] = None
        best_feature["variance_reduced"] = 0
        
        maximum_variance_reduction = -float("inf")
        
        for features in range(number_attributes):
            values = dataset[:, features]
            unique_sorted_values = np.unique(values)
            threshold_array = np.array([(unique_sorted_values[i]+unique_sorted_values[i+1])/2 for i in range(0,len(unique_sorted_values)-1)])

            for threshold in threshold_array:
                dataset_left, dataset_right = self.split_left_right(dataset, features, threshold)
                
                if len(dataset_left)>0 and len(dataset_right)>0:
                    
                    dataset_y, dataset_left_y, dataset_right_y = dataset[:, -1], dataset_left[:, -1], dataset_right[:, -1]
                    
                    variance_reduced = self.variance_reduction(dataset_y, dataset_left_y, dataset_right_y)
                    if variance_reduced > maximum_variance_reduction:
                        best_feature["attribute"] = features
                        best_feature["threshold"] = threshold
                        best_feature["dataset_left"] = dataset_left
                        best_feature["dataset_right"] = dataset_right
                        best_feature["variance_reduced"] = variance_reduced
                        maximum_variance_reduction = variance_reduced
                        
        return best_feature
    
    def split_left_right(self, dataset, index, threshold):
        
        left_dataset = np.array([x for x in dataset if x[index]<=threshold])
        right_dataset = np.array([x for x in dataset if x[index]>threshold])
        return left_dataset, right_dataset
    
    def variance_reduction(self, parent, left_branch, right_branch):
        
        fraction_left = len(left_branch) / len(parent)
        fraction_right = len(right_branch) / len(parent)
        reduction_variance = np.var(parent) - (fraction_left * np.var(left_branch) + fraction_right * np.var(right_branch))
        return reduction_variance
    
    def cal_leaf_node(self, y):
        
        leaf_val = np.mean(y)
        return leaf_val
                
    def print_decision_tree(self,columns,decision_tree=None,indent=" "):
        
        if not decision_tree:
            decision_tree = self.root

        if decision_tree.leaf_value is not None:
            print("Leaf: ",round(decision_tree.leaf_value,3))

        else:
            print(columns[decision_tree.attribute], "==>", round(decision_tree.threshold,3), "(", round(decision_tree.variance_red,3),")")
            print("%sLeft: " % (indent), end="")
            self.print_decision_tree(columns, decision_tree.child_left, indent+indent)
            print("%sRight: " % (indent), end="")
            self.print_decision_tree(columns, decision_tree.child_right, indent+indent)
    
    def fit_model(self, X, y):
        
        dataset = np.concatenate((X, y), axis=1)
        self.root = self.construct_tree(dataset)
        
    def predict(self, data, decision_tree=None):
        
        if not decision_tree:
            decision_tree = self.root
            
        if decision_tree.leaf_value!=None: 
            return decision_tree.leaf_value
        
        attribute_value = data[decision_tree.attribute]
        if attribute_value <= decision_tree.threshold:
            return self.predict(data, decision_tree.child_left)
        else:
            return self.predict(data, decision_tree.child_right)
    
```

```python
def mean_error(y_pred, y_actual, n):
    sum=0
    for i in range(n):
        sum = sum+(y_pred[i]-y_actual[i])**2
    sum = sum/n
    sum = np.sqrt(sum)
    return sum
```

```python
max_error=float("inf")
for i in range(10):
    d = data.sample(frac = 1,random_state=42)
    div = int(0.7 * d.shape[0])
    d_train, d_test = d.iloc[:div,:], d.iloc[div:,:]
    d_train_x = d_train.iloc[:,:-1].values
    d_train_y = d_train.iloc[:,-1].values.reshape(-1,1)
    d_test_x = d_test.iloc[:,:-1].values
    d_test_y = d_test.iloc[:,-1].values.reshape(-1,1)
    regress_tree = RegressionTree(minimum_samples=3, max_depth=20)
    regress_tree.fit_model(d_train_x,d_train_y)
    y_pred_train = [regress_tree.predict(x,regress_tree.root) for x in d_train_x]
    if mean_error(y_pred_train,d_train_y,d_train_x.shape[0])<max_error:
        max_error = mean_error(y_pred_train,d_train_y,d_train_x.shape[0])
        dataset_train = d_train
        dataset_test = d_test
        
```

```python
train = []
test = []
for i in range(3,20):
    regress_tree = RegressionTree(minimum_samples=3, max_depth=i)
    regress_tree.fit_model(data_train_x,data_train_y)
    
    y_pred_train = [regress_tree.predict(x,regress_tree.root) for x in data_train_x] 
    train.append(mean_error(y_pred_train,data_train_y,data_train_x.shape[0]))
    
    y_pred_test = [regress_tree.predict(x,regress_tree.root) for x in data_test_x] 
    test.append(mean_error(y_pred_test,data_test_y,data_test_x.shape[0]))

```

```python
import copy
def post_pruning(decision_tree,dataset,regress_tree,error=test[-1]):
        root = regress_tree.root
        tree1 = copy.copy(decision_tree)
        tree2 = copy.deepcopy(decision_tree)
        if tree1.leaf_value is None:
            tree1.left_child = None
            tree1.right_child = None
            tree1.leaf_value = regress_tree.cal_leaf_node(y)
            y_pred = [regress_tree.predict(x,root) for x in X] 
            if (mean_error(y_pred,y,X.shape[0]))<error:
                tree2 = copy.copy(tree1)
                error = mean_error(y_pred,y,X.shape[0])
                return root
            else:
                tree1 = copy.copy(tree2)
                left = [row for row in dataset if row[tree1.attribute]<=tree1.threshold]
                right = [row for row in dataset if row[tree1.attribute]>tree1.threshold]
                a = post_pruning(tree1.child_left,left,error)
                b = post_pruning(tree1.child_right,right,error)
                return Node(root.attribute, root.threshold,a,b,root.variance_red)
```

```python
tree = copy.deepcopy(regress_tree.root)
X = dataset_test.iloc[:,:-1].values
y = dataset_test.iloc[:,-1].values.reshape(-1,1)
dataset = np.concatenate((X, y), axis=1)
pruned = post_pruning(tree,dataset,regress_tree)

y_pred_test = [regress_tree.predict(x,pruned) for x in data_test_x] 
mean_error(y_pred_test,data_test_y,data_test_x.shape[0])
```
