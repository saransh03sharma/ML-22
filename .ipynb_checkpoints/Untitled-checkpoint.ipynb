```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv("Train_B_Tree.csv")
```

```python
data = data.drop_duplicates(keep='first')
```

```python
X = data.drop(['csMPa'],axis=1)
X.shape
```

```python
y = data.csMPa
y.shape
```

```python
data = (X - X.mean()) / (X.std())              # standardization
data = pd.concat([y,data],axis=1)
data = pd.melt(data,id_vars="csMPa",
                    var_name="features",
                    value_name='value')
plt.figure(figsize=(10,10))
sns.violinplot(x="features", y="value", data=data, inner="quart")
plt.xticks(rotation=45);
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(x="features", y="value", data=data)
plt.xticks(rotation=45);
```
